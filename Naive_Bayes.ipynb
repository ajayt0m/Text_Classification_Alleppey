```python
import pandas as pd

file_path = r"https://raw.githubusercontent.com/ajayt0m/Text_Classification_Alleppey/main/WScraped.csv"

# Try reading the CSV file with 'latin1' encoding
df = pd.read_csv(file_path, encoding='latin1')
```

```python
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df)
```

```python
# Fill NaN values in 'Title' column with an empty string
df['Title'].fillna('', inplace=True)

# Create a new column 'Title_Review' by concatenating 'Title' and 'Review'
df['Title_Review'] = df['Title'] + '. ' + df['Review']
```

```python
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df)
```

# Case Normalizing

```python
df['Title_Review'] = df['Title_Review'].str.lower()
```

```python
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df.head())
```

```python
df['Title_Review'].isnull().sum()
```

```python
df['Label'].value_counts()
```

```python
# Define a function to simplify the labels
def simplify_label(label):
    if label in [4, 5]:
        return 'Happy Visit'
    elif label == 3:
        return 'Ok Visit'
    elif label in [1, 2]:
        return 'Bad Visit'
    else:
        return 'Unknown'  # You can adjust this based on your specific needs

# Apply the function to create the new column 'Label_simplified'
df['Label_simplified'] = df['Label'].apply(simplify_label)
```

```python
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df.head())
```

# Selecting Features

```python
# Create Feature and Label sets
df = df[['Title_Review','Label', 'Label_simplified']]
```

# Train - Test Splitting

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.33, random_state=42)

print('Training Data Shape:', train.shape)
print('Testing Data Shape: ', test.shape)
```

```python
#Importing necessary libraires
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

# Assigning Weights to avoid bias

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Compute class weights based on the training data
class_weights = compute_class_weight('balanced', classes=np.unique(train['Label_simplified']), y=train['Label_simplified'])

# Create a dictionary to pass to the model
class_weight_dict = dict(zip(np.unique(train['Label_simplified']), class_weights))
```

```python
class_weight_dict
```

```python
TfidfVectorizer(train['Title_Review'])
```

```python
# Create a model based on Multinomial Naive Bayes with class weights
model = make_pipeline(TfidfVectorizer(), MultinomialNB(fit_prior=False, class_prior=class_weights))

# Train the model with the training data
model.fit(train['Title_Review'], train['Label_simplified'])

# Predict on the test data
predicted = model.predict(test['Title_Review'])
```

```python
predicted
```

```python
# Get unique class labels
class_labels = np.unique(test['Label_simplified'])
```

```python
#Creating confusion Matrix and heatmap
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.Label_simplified, predicted)

import seaborn as sns
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False
            , xticklabels=class_labels,
            yticklabels=class_labels)

#Plotting Heatmap of confusion Matrix
import matplotlib.pyplot as plt
plt.xlabel('true label')
plt.ylabel('predicted label')
```
